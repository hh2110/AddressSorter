```python
#try k-means clustering algorithm
# rather than calculating euclidena distance 
# betweeen addresses, we will use difflib package
# to see difference - and can cluster accordingly.

#We will create 'our' own K-means medoid code.
# http://www.math.le.ac.uk/people/ag153/homepage/KmeansKmedoids/Kmeans_Kmedoids.html
# algorithm:
# 1. Initialize: randomly select k of the n 
#                data points as the medoids
# 2. Assignment step: Associate each data point 
#                     to the closest medoid.
# 3. Update step: For each medoid m and each data 
#         point o associated to m swap m and o and 
#         compute the total cost of the configuration 
#         (that is, the average dissimilarity of o to 
#         all the data points associated to m). Select 
#         the medoid o with the lowest cost of the 
#       configuration.

#need to consider the fact that we have multiple
        # and spaces in both eachAddress and eachMed
        # but could just remove the spaces and see what happens
        # instead....
```

```python
def CalcMatchPAvg(listOfAddresses, medoidName):
    matchPL=[]
    for i in listOfAddresses:
        matchP=difflib.SequenceMatcher(isjunk=None,
                                     a=i,
                                     b=medoidName)
        matchP=matchP.ratio()*100
        matchPL.append(matchP)
    return(np.mean(matchPL))

def AssignmnetStep(medoidList, address_df):
    newChosenMedoids=[]
    chosenMedoidMatchP=[]
    #choose address in df
    for eachAddress in address_df['Address']:
        matchPL=[]
        #choose medoid
        for eachMed in medoidList:
            #calc match% between medoid and address
            matchP=difflib.SequenceMatcher(isjunk=None,
                                         a=eachMed,
                                         b=eachAddress)
            matchP=matchP.ratio()*100
            matchPL.append(matchP)
        #get the maximum match and assign the medoid
        chosenMedoid=medoidList[matchPL.index(max(matchPL))]
        chosenMedoidMatchP.append(max(matchPL))
        newChosenMedoids.append(chosenMedoid)

    #add in column for medoids into df
    address_df['medoid']=pd.Series(newChosenMedoids, index=address_df.index)
    address_df['matchP']=pd.Series(chosenMedoidMatchP, index=address_df.index)
    return address_df
```
